```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, classification_report


df = pd.read_csv("/content/drive/MyDrive/archive/merged_weather_data.csv")


df["Rain (last 1 hour)"] = df["Rain (last 1 hour)"].fillna(0)
df["Rain (last 1 hour)"] = df["Rain (last 1 hour)"].clip(upper=10)

def categorize_rain(mm):
    if mm == 0:
        return 0  # No rain
    elif mm <= 1:
        return 1  # Light rain
    elif mm <= 5:
        return 2  # Moderate rain
    else:
        return 3  # Heavy rain


df["Rain_Category"] = df["Rain (last 1 hour)"].apply(categorize_rain)


X = df.drop(columns=["Rain (last 1 hour)", "Rain_Category", "Visibility", "Date", "City"])
y_cls = df["Rain_Category"]
y_reg = df["Rain (last 1 hour)"]


categorical_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()


numerical_transformer = SimpleImputer(strategy="mean")
categorical_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),("onehot", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", numerical_transformer, numerical_cols),("cat", categorical_transformer, categorical_cols)
])


X_train, X_test, y_train_cls, y_test_cls = train_test_split(X, y_cls, test_size=0.2, random_state=42)
_, _, y_train_reg, y_test_reg = train_test_split(X, y_reg, test_size=0.2, random_state=42)


X_train_proc = preprocessor.fit_transform(X_train)
X_test_proc = preprocessor.transform(X_test)


clf_multi = RandomForestClassifier(n_estimators=100, random_state=42)
clf_multi.fit(X_train_proc, y_train_cls)
cls_pred = clf_multi.predict(X_test_proc)


regressors = {}
for category in [1, 2, 3]:
    mask = y_train_cls == category
    reg = Ridge()
    reg.fit(X_train_proc[mask], y_train_reg[mask])
    regressors[category] = reg


final_pred = np.zeros_like(cls_pred, dtype=float)
for category, model in regressors.items():
    mask = cls_pred == category
    final_pred[mask] = model.predict(X_test_proc[mask])

final_pred = np.clip(final_pred, 0, None)


cls_report = classification_report(y_test_cls, cls_pred)
mae = mean_absolute_error(y_test_reg, final_pred)
rmse = np.sqrt(mean_squared_error(y_test_reg, final_pred))
r2 = r2_score(y_test_reg, final_pred)

print("Classification Report:\n", cls_report)
print(f"\nMAE: {mae:.3f} ")
print(f"RMSE: {rmse:.3f} ")
print(f"R^2 Score: {r2:.3f}")


X_full_proc = preprocessor.transform(X)
cls_full_pred = clf_multi.predict(X_full_proc)
final_full_pred = np.zeros_like(cls_full_pred, dtype=float)
for category, model in regressors.items():
    mask = cls_full_pred == category
    final_full_pred[mask] = model.predict(X_full_proc[mask])
final_full_pred = np.clip(final_full_pred, 0, None)

df["Predicted_Rain_1hr"] = final_full_pred
output_file = "rainfall_with_multiclass_predictions.csv"
df.to_csv(output_file, index=False)
print(f"\nPredictions saved to: {output_file}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay,
    roc_curve, auc
)
from imblearn.over_sampling import SMOTE


file_path = "/content/drive/MyDrive/archive/updated_rainfall_flood_dataset.csv"
df = pd.read_csv(file_path)


df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.month


df = df.drop(columns=["Date", "Visibility"], errors="ignore")


label_encoders = {}
for col in ["City", "Weather_Description"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le


X = df.drop(columns=["Flood_Occurrence"])
y = df["Flood_Occurrence"]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train_scaled, y_train)

models = {
    "Random Forest": RandomForestClassifier(n_estimators=200, max_depth=15, random_state=42),
    "XGBoost": XGBClassifier(eval_metric='logloss', max_depth=6, n_estimators=150, random_state=42)
}


best_model = None
best_score = 0

for name, model in models.items():
    model.fit(X_train_res, y_train_res)
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_pred)

    print(f"\n{name} Performance:")
    print(f"Accuracy : {acc:.4f}")
    print(f"F1 Score : {f1:.4f}")
    print(f"ROC AUC  : {roc:.4f}")

    if acc > best_score:
        best_score = acc
        best_model = model


X_scaled = scaler.transform(X)
df["Predicted_Flood_Occurrence"] = best_model.predict(X_scaled)
df["Flood_Probability"] = best_model.predict_proba(X_scaled)[:, 1]
df["City_Name"] = label_encoders["City"].inverse_transform(df["City"])

city_input = input("Enter a city to show monthly average rainfall: ").strip().title()


if city_input in label_encoders["City"].classes_:
    city_encoded = label_encoders["City"].transform([city_input])[0]
    monthly_rainfall = df[df["City"] == city_encoded].groupby("Month")["Predicted_Rain_1hr"].mean()

    plt.figure(figsize=(10, 5))
    plt.plot(monthly_rainfall.index, monthly_rainfall.values, marker='o', color='dodgerblue')
    plt.title(f"Monthly Average Rainfall in {city_input}")
    plt.xlabel("Month")
    plt.ylabel("Average Rainfall (mm)")
    plt.grid(True)
    plt.xticks(np.arange(1, 13))
    plt.tight_layout()
    plt.show()
else:
    print(f"City '{city_input}' not found in dataset.")

y_prob = best_model.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)


df.to_csv("flood_predictions_improved.csv", index=False)
print("\n Predictions with confidence scores saved to 'flood_predictions_improved.csv'")
```
